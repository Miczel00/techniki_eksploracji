```python
#7. Przeprowadzić transponowanie macierzy ‘na piechotę’.
rez = [[W[j][i] for j in range(len(W))] for i in range(len(W[0]))]
transpozed_matrix = [row for row in rez]
print(np.array(transpozed_matrix))
```
